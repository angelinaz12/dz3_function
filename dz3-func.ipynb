```python
"""Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class"""

import pandas as pd
```

```python
df = pd.read_csv('ratings.csv')
df.head()
```

```python
"функция,которая считает класс фильма в зависимсоти от рейтинга"

def movie_class(i):
    if i <= 2:
        return 'low'
    if i <= 4 and i >= 2.1:
        return 'medium'
    if i >= 4.5 and i <= 5:
        return 'high'
```

```python
"создаем новый столбец class и записываем в него результат работы функции, которую применяем к столбцу rating"
df['class'] = df['rating'].apply(movie_class)
df.head(10)
```

```python
"""Используем файл keywords.csv.
Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую 
принадлежность определенному региону. 
Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона.
Если поисковый запрос не содержит названия города, то ставим ‘undefined’. 
Результат классификации запишите в отдельный столбец region."""

geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо - Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

keywords = pd.read_csv('keywords.csv')
keywords

```

```python
"""функция, которая выставляет гео-классификатор в зависимсоти от того содержится ли название города в запросе"""
def geo_teg(row):
    for key, value in geo_data.items():
        for i in value:
            if i in row['keyword']:
                return key
        return 'undefined'
        
    
```

```python
#применяем функцию к столбцу keyword и записывает рузультат в новый столбец regoin 
keywords['region'] = keywords.apply(geo_teg, axis = 1)
#вылитруем по столбцу regoin,чтобы проверить результат работы функции
keywords[(keywords['region'] == 'Центр')].head(15)
```
